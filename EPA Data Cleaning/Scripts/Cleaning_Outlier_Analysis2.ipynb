```python

import numpy as np
import pandas as pd
from scipy.stats import trim_mean
```

```python
import pandas as pd

data = pd.read_excel(r"C:\Users\clair\Senior-Project\EPA Data Cleaning\Raw Data\Excel\ExtraEPAData_diesel.xlsx")
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
df2 = df.rename(columns={"Unnamed: 0":"label"})
df2.head()
```

```python

df2["vehicle_group"] = df2["label"].where(
    df2["label"].str.contains("Buses|Heavy-duty", na=False)
).ffill()
```

```python
df2.head(10)
```

```python
co2_rows = df2[df2["label"] == "Exhaust CO2"]
```

```python
co2_rows.head(10)
```

```python
year_cols = co2_rows.columns.difference(["label", "vehicle_group"])

bus_values = co2_rows.loc[
    co2_rows["vehicle_group"].str.contains("Buses"),
    year_cols
].to_numpy().flatten()

heavy_values = co2_rows.loc[
    co2_rows["vehicle_group"].str.contains("Heavy-duty"),
    year_cols
].to_numpy().flatten()
```

```python
large_values = np.concatenate([bus_values, heavy_values])

large_co2 = trim_mean(large_values, 0.1)  # 10% trim
```

```python
large_df = pd.DataFrame({
    "Vehicle Type": ["Large"],
    "CO2 (g/mi)": [round(large_co2, 4)]
})

large_df.head()
```

```python
data2 = pd.read_csv("C:/Users/clair/Senior-Project/EPA Data Cleaning/Clean_outlier/carbon_emissions.csv")
df2 = pd.DataFrame(data2)
df2.head()
```

```python
merged = pd.concat([large_df,df2],ignore_index=True)
merged.head()
```

```python
merged.to_csv('carbon_emissions_final.csv')
```
