Convert data to dataframes to begin cleaning

```python
df_25 = pd.DataFrame(data_25)
df_24 = pd.DataFrame(data_24)
df_23 = pd.DataFrame(data_23)
df_22 = pd.DataFrame(data_22)
df_21 = pd.DataFrame(data_21)
df_20 = pd.DataFrame(data_20)
df_19 = pd.DataFrame(data_19)
df_18 = pd.DataFrame(data_18)
df_17 = pd.DataFrame(data_17)
df_16 = pd.DataFrame(data_16)
df_15 = pd.DataFrame(data_15)
df_14 = pd.DataFrame(data_14)
df_13 = pd.DataFrame(data_13)
df_12 = pd.DataFrame(data_12)
df_11 = pd.DataFrame(data_11)
df_10 = pd.DataFrame(data_10)
df_09lr = pd.DataFrame(data_09lr)
```

Append dataframes for 2017-2025 into one dataframe, and 2009-2016 into one dataframe

```python
merged_17to25 = pd.concat([df_25,df_24,df_23,df_22,df_21,df_20,df_19,df_18,df_17], ignore_index=True)
merged_09to16 = pd.concat([df_16,df_15,df_14,df_13,df_12,df_11,df_10,df_09lr], ignore_index=True)
```

Filter out unnecessary columns

```python
columns_to_keep = ['Veh Mfr Code', 'Represented Test Veh Make', 'Represented Test Veh Model', 'Test Vehicle ID', 'Vehicle Type', 'Test Fuel Type Cd', 'Test Fuel Type Description', 'CO2 (g/mi)']
filtered_17to25 = merged_17to25[columns_to_keep]
filtered_09to16 = merged_09to16[columns_to_keep]
```

Average CO2 emissions for cars of same make and model from different years

```python
avg_17to25 = filtered_17to25.groupby(['Represented Test Veh Make', 'Represented Test Veh Model']).agg({
    'CO2 (g/mi)': 'mean',
    'Veh Mfr Code': 'first', 
    'Vehicle Type': 'first',
    'Test Vehicle ID': 'first'
}).reset_index()

avg_09to16 = filtered_09to16.groupby(['Represented Test Veh Make', 'Represented Test Veh Model']).agg({
    'CO2 (g/mi)': 'mean',
    'Veh Mfr Code': 'first',  
    'Vehicle Type': 'first',
    'Test Vehicle ID': 'first'
}).reset_index()
```

Drop rows that do not report any CO2 emissions

```python
clean_17to25 = avg_17to25.dropna(subset=['CO2 (g/mi)'])
clean_09to16 = avg_09to16.dropna(subset=['CO2 (g/mi)'])
```
