```python
merged_17to25 = pd.concat([df_25,df_24,df_23,df_22,df_21,df_20,df_19,df_18,df_17], ignore_index=True)
merged_09to16 = pd.concat([df_16,df_15,df_14,df_13,df_12,df_11,df_10,df_09lr], ignore_index=True)
```

```python
merged = pd.concat([merged_09to16,merged_17to25],ignore_index=True)
```

```python
columns_to_keep = ['Veh Mfr Code', 'Represented Test Veh Make', 'Represented Test Veh Model',
                'Test Vehicle ID', 'Vehicle Type', 'Test Fuel Type Cd', 'Test Fuel Type Description',
                 'CO2 (g/mi)']
filtered = merged[columns_to_keep]
```

```python
clean = filtered.dropna(subset=['CO2 (g/mi)'])
```

```python

from scipy.stats import trim_mean

co2_by_size = (
    clean.groupby("Vehicle Type")["CO2 (g/mi)"]
      .apply(lambda x: trim_mean(x, 0.1))
)

co2_by_size.head(10)
```

```python
co2_by_size = co2_by_size.drop("Both")
```

```python
co2_by_size = co2_by_size.rename({"small": "Small"})
co2_by_size = co2_by_size.rename({"Truck":"Medium"})
co2_by_size.head()
```
